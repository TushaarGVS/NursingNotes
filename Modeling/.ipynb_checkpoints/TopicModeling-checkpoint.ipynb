```python
import pandas as pd
import numpy as np
import operator

import gensim
from gensim import corpora, models

import IPython
import pprint as pp
```

```python
tf_idfMIMIC = pd.read_csv('../Dataset/MIMIC_TFIDF.csv', header=None)
embeddingMIMIC_500 = pd.read_csv('../Dataset/MIMIC_EMBEDDING_500.csv', header=None)
embeddingMIMIC_1000 = pd.read_csv('../Dataset/MIMIC_EMBEDDING_1000.csv', header=None)

IPython.display.display(tf_idfMIMIC.head())
print tf_idfMIMIC.shape
IPython.display.display(embeddingMIMIC_500.head())
print embeddingMIMIC_500.shape
IPython.display.display(embeddingMIMIC_1000.head())
print embeddingMIMIC_1000.shape
```

```python
preprocessedMIMIC = pd.read_csv("../Dataset/Preprocessed_MIMIC.csv")
IPython.display.display(preprocessedMIMIC.head())
print preprocessedMIMIC.shape
```

```python
preprocessedTextMIMIC = preprocessedMIMIC['PREPROCESSED_TEXT'].str.split(" ") 
IPython.display.display(preprocessedTextMIMIC.head())
```

```python
wordDictionary = gensim.corpora.Dictionary(preprocessedTextMIMIC)
idx = 0
for count, term in wordDictionary.iteritems():
    if idx < 10:
        print(count, term)
        idx += 1
wordDictionary.filter_extremes(no_below=10)
```

```python
bagOfWordsCorpus = [wordDictionary.doc2bow(tokensList) for tokensList in preprocessedTextMIMIC]
pp.pprint(bagOfWordsCorpus[0])

tfIdfModel = models.TfidfModel(bagOfWordsCorpus)
tfIdfCorpus = tfIdfModel[bagOfWordsCorpus]
pp.pprint(tfIdfCorpus[0])
```

```python
# truncation of 150 by default
bagOfWordsHDP = gensim.models.HdpModel(corpus=bagOfWordsCorpus, id2word=wordDictionary)
tfIdfHDP = gensim.models.HdpModel(corpus=tfIdfCorpus, id2word=wordDictionary)
```

```python
bagOfWordsHDP.optimal_ordering()
tfIdfHDP.optimal_ordering()

bagOfWordsHDP.save(fname_or_handle="models/bagOfWordsHDP")
tfIdfHDP.save(fname_or_handle="models/tfIdfHDP")
```

```python
pp.pprint(bagOfWordsHDP.print_topics(num_topics=5, num_words=10))
pp.pprint(tfIdfHDP.print_topics(num_topics=5, num_words=10))
```

```python
def LDAasLSI(threshold=0.7, maxIterns=5):
    topTopics = [(0, 0)]
    while topTopics[0][1] < threshold and maxIterns > 0:
        bagOfWordsLDA = gensim.models.LdaModel(corpus=bagOfWordsCorpus, id2word=wordDictionary)
        coherenceValues = {}
        for idx, topic in bagOfWordsLDA.show_topics(num_topics=-1, formatted=False):
            topic = [token for token, score in topic]
            coherenceModel = gensim.models.CoherenceModel(topics=[topic], window_size=10, dictionary=wordDictionary, texts=preprocessedTextMIMIC)
            coherenceValues[idx] = coherenceModel.get_coherence()
        topTopics = sorted(coherenceValues.items(), key=operator.itemgetter(1), reverse=True)
        maxIterns -= 1
    return bagOfWordsLDA, topTopics, threshold

bagOfWordsLDA, topTopicsLDA, threshold = LDAasLSI(0.9, 10)
bagOfWordsLDA.save(fname='LDAasLSI')
print threshold
print topTopicsLDA[:5]
```

```python
def generateDatasets(corpus, model, numClusters, datasetName):
    clusterList = []
    for idx in range(len(preprocessedTextMIMIC)):
        textClusters = [0] * numClusters
        clusters = model[corpus[idx]]
        for clusterNumber, probability in clusters:
            textClusters[clusterNumber] = probability
        textClusters.append(preprocessedMIMIC.iloc[idx, 6])
        clusterList.append(textClusters)
    pd.DataFrame(clusterList).to_csv(datasetName, index=False)
    
generateDatasets(corpus=bagOfWordsCorpus, model=bagOfWordsHDP, numClusters=150, datasetName='../Dataset/bagOfWordsHDP.csv')
generateDatasets(corpus=tfIdfCorpus, model=tfIdfHDP, numClusters=150, datasetName='../Dataset/tfIdfHDP.csv')
generateDatasets(corpus=bagOfWordsCorpus, model=bagOfWordsLDA, numClusters=100, datasetName='../Dataset/bagOfWordsLDAasLSI.csv')
```
