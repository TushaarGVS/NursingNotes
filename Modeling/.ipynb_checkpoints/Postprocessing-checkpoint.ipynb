```python
import pandas as pd
import numpy as np

import IPython
import pprint as pp
```

```python
tfIdfMIMIC = pd.read_csv('../Dataset/MIMIC_TFIDF.csv', header=None)
embeddingMIMIC_500 = pd.read_csv('../Dataset/MIMIC_EMBEDDING_500.csv', header=None)
embeddingMIMIC_1000 = pd.read_csv('../Dataset/MIMIC_EMBEDDING_1000.csv', header=None)
bagOfWordsHDP = pd.read_csv('../Dataset/bagOfWordsHDP.csv')
tfIdfHDP = pd.read_csv('../Dataset/tfIdfHDP.csv')
bagOfWordsLDAasLSI = pd.read_csv('../Dataset/bagOfWordsLDAasLSI.csv')

IPython.display.display(tfIdfMIMIC.head())
print tfIdfMIMIC.shape
IPython.display.display(embeddingMIMIC_500.head())
print embeddingMIMIC_500.shape
IPython.display.display(embeddingMIMIC_1000.head())
print embeddingMIMIC_1000.shape
IPython.display.display(bagOfWordsHDP.head())
print bagOfWordsHDP.shape
IPython.display.display(tfIdfHDP.head())
print tfIdfHDP.shape
IPython.display.display(bagOfWordsLDAasLSI.head())
print bagOfWordsLDAasLSI.shape
```

```python
# removing error entries
errorEntries = [395, 2102, 2411, 4673, 5123, 5988, 6632, 7765, 7876]
tfIdfMIMIC = tfIdfMIMIC.drop(errorEntries)
embeddingMIMIC_500 = embeddingMIMIC_500.drop(errorEntries)
embeddingMIMIC_1000 = embeddingMIMIC_1000.drop(errorEntries)
bagOfWordsHDP = bagOfWordsHDP.drop(errorEntries)
tfIdfHDP = tfIdfHDP.drop(errorEntries)
bagOfWordsLDAasLSI = bagOfWordsLDAasLSI.drop(errorEntries)
```

```python
def classMapping(dataframe):
    mapping = []
    for idx in range(dataframe.shape[0]):
        mapping.append(map(int, dataframe.iloc[idx, -1][1:-1].split(',')))
    dataframe = dataframe.drop(dataframe.columns[-1], axis=1)
    dataframe['class'] = mapping
    return dataframe

tfIdfMIMIC = classMapping(tfIdfMIMIC)
embeddingMIMIC_500 = classMapping(embeddingMIMIC_500)
embeddingMIMIC_1000 = classMapping(embeddingMIMIC_1000)
bagOfWordsHDP = classMapping(bagOfWordsHDP)
tfIdfHDP = classMapping(tfIdfHDP)
bagOfWordsLDAasLSI = classMapping(bagOfWordsLDAasLSI)
```

```python
tfIdfMIMIC.to_csv('../FinalDataset/tfIdfMIMIC.csv', index=False)
embeddingMIMIC_500.to_csv('../FinalDataset/embeddingMIMIC_500.csv', index=False)
embeddingMIMIC_1000.to_csv('../FinalDataset/embeddingMIMIC_1000.csv', index=False)
bagOfWordsHDP.to_csv('../FinalDataset/bagOfWordsHDP.csv', index=False)
tfIdfHDP.to_csv('../FinalDataset/tfIdfHDP.csv', index=False)
bagOfWordsLDAasLSI.to_csv('../FinalDataset/bagOfWordsLDAasLSI.csv', index=False)
```
