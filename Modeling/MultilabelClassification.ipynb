```python
def classMapping(dataframe):
    mapping = []
    for idx in range(dataframe.shape[0]):
        mapping.append(map(int, dataframe.iloc[idx, -1][1:-1].split(',')))
    dataframe = dataframe.drop(dataframe.columns[-1], axis=1)
    dataframe['class'] = mapping
    return dataframe

# tfIdfMIMIC = classMapping(tfIdfMIMIC)
embeddingMIMIC_500 = classMapping(embeddingMIMIC_500)
embeddingMIMIC_1000 = classMapping(embeddingMIMIC_1000)
bagOfWordsHDP = classMapping(bagOfWordsHDP)
tfIdfHDP = classMapping(tfIdfHDP)
bagOfWordsLDAasLSI = classMapping(bagOfWordsLDAasLSI)
```

```python
def average(inputList):
    return sum(inputList)/float(len(inputList))
```

```python
def evaluation(datasets, numFolds=10):
    overallEvaluation = {}
    for dataset in datasets:
        datasetEvaluation = {}
        print "Dataset: ", dataset.name
        coverageErrors, avgPrecisionRankings, rankingLosses = {'randomForest': [], 'multilayerPerceptron': [], 'knn': []}, {'randomForest': [], 'multilayerPerceptron': [], 'knn': []}, {'randomForest': [], 'multilayerPerceptron': [], 'knn': []}
        kFolds = KFold(n_splits=numFolds, shuffle=True, random_state=8)
        fold = 1
        for result in kFolds.split(dataset):
            trainDf = dataset.iloc[result[0]]
            testDf = dataset.iloc[result[1]]
            X_train, y_train = trainDf.drop(trainDf.columns[-1], axis=1).values, np.array([row for row in trainDf.iloc[:, -1].values])
            X_validate, y_validate = testDf.drop(testDf.columns[-1], axis=1).values, np.array([row for row in testDf.iloc[:, -1].values])
            print "\nFold-%s is split" %fold
            
            randomForest.fit(X_train, y_train)
            multilayerPerceptron.fit(X_train, y_train)
            knn.fit(X_train, y_train)
            print "RF, MLP and KNN are built"
            
            predictedRF = randomForest.predict(X_validate).astype(int).tolist()
            predictedMLP = multilayerPerceptron.predict(X_validate).astype(int).tolist()
            predictedKNN = knn.predict(X_validate).astype(int).tolist()
            print "All model predictions are made"
            
            coverageErrors['randomForest'].append(coverage_error(y_true=y_validate, y_score=predictedRF))
            coverageErrors['multilayerPerceptron'].append(coverage_error(y_true=y_validate, y_score=predictedMLP))
            coverageErrors['knn'].append(coverage_error(y_true=y_validate, y_score=predictedKNN))
            
            avgPrecisionRankings['randomForest'].append(label_ranking_average_precision_score(y_true=y_validate, y_score=predictedRF))
            avgPrecisionRankings['multilayerPerceptron'].append(label_ranking_average_precision_score(y_true=y_validate, y_score=predictedMLP))
            avgPrecisionRankings['knn'].append(label_ranking_average_precision_score(y_true=y_validate, y_score=predictedKNN))
            
            rankingLosses['randomForest'].append(label_ranking_loss(y_true=y_validate, y_score=predictedRF))
            rankingLosses['multilayerPerceptron'].append(label_ranking_loss(y_true=y_validate, y_score=predictedMLP))
            rankingLosses['knn'].append(label_ranking_loss(y_true=y_validate, y_score=predictedKNN))
            print "All evaluation metrics are computed"
            fold += 1
        print "\n", "-" * 117
        print "Coverage errors"
        pp.pprint(coverageErrors)
        print "\nLabel Ranking Avg. Precision"
        pp.pprint(avgPrecisionRankings)
        print "\nLabel Ranking Loss"
        pp.pprint(rankingLosses)
        print "-" * 117, "\n"
        
        datasetEvaluation['randomForest'] = {'coverageError': average(coverageErrors['randomForest']), 'labelRankingAvgPrecision': average(avgPrecisionRankings['randomForest']), 'labelRankingLoss': average(rankingLosses['randomForest'])}
        datasetEvaluation['multilayerPerceptron'] = {'coverageError': average(coverageErrors['multilayerPerceptron']), 'labelRankingAvgPrecision': average(avgPrecisionRankings['multilayerPerceptron']), 'labelRankingLoss': average(rankingLosses['multilayerPerceptron'])}
        datasetEvaluation['knn'] = {'coverageError': average(coverageErrors['knn']), 'labelRankingAvgPrecision': average(avgPrecisionRankings['knn']), 'labelRankingLoss': average(rankingLosses['knn'])}
        avgCoverageError = (datasetEvaluation['randomForest']['coverageError'] + datasetEvaluation['multilayerPerceptron']['coverageError'] + datasetEvaluation['knn']['coverageError'])/float(3)
        avgAvgPrecisionRanking = (datasetEvaluation['randomForest']['labelRankingAvgPrecision'] + datasetEvaluation['multilayerPerceptron']['labelRankingAvgPrecision'] + datasetEvaluation['knn']['labelRankingAvgPrecision'])/float(3)
        avgRankingLoss = (datasetEvaluation['randomForest']['labelRankingLoss'] + datasetEvaluation['multilayerPerceptron']['labelRankingLoss'] + datasetEvaluation['knn']['labelRankingLoss'])/float(3)
        datasetEvaluation['average'] = {'coverageError': avgCoverageError, 'labelRankingAvgPrecision': avgAvgPrecisionRanking, 'labelRankingLoss': avgRankingLoss}
        overallEvaluation[dataset.name] = datasetEvaluation
    return overallEvaluation
```

```python
def evaluationWithOtherModel(datasets, model, numFolds=10):
    overallEvaluation = {}
    for dataset in datasets:
        datasetEvaluation = {}
        print "Dataset: ", dataset.name
        coverageErrors, avgPrecisionRankings, rankingLosses = [], [], []
        kFolds = KFold(n_splits=numFolds, shuffle=True, random_state=8)
        fold = 1
        for result in kFolds.split(dataset):
            trainDf = dataset.iloc[result[0]]
            testDf = dataset.iloc[result[1]]
            X_train, y_train = trainDf.drop(trainDf.columns[-1], axis=1).values, np.array([row for row in trainDf.iloc[:, -1].values])
            X_validate, y_validate = testDf.drop(testDf.columns[-1], axis=1).values, np.array([row for row in testDf.iloc[:, -1].values])
            print "\nFold-%s is split" %fold
            
            model.fit(X_train, y_train)
            print "Model is built"
            
            predicted = model.predict(X_validate).astype(int).tolist()
            print "Model predictions are made"
            
            coverageErrors.append(coverage_error(y_true=y_validate, y_score=predicted))
            avgPrecisionRankings.append(label_ranking_average_precision_score(y_true=y_validate, y_score=predicted))
            rankingLosses.append(label_ranking_loss(y_true=y_validate, y_score=predicted))
            print "All evaluation metrics are computed"
            fold += 1
        print "\n", "-" * 117
        print "Coverage errors"
        pp.pprint(coverageErrors)
        print "\nLabel Ranking Avg. Precision"
        pp.pprint(avgPrecisionRankings)
        print "\nLabel Ranking Loss"
        pp.pprint(rankingLosses)
        print "-" * 117, "\n"
        
        avgCoverageError = average(coverageErrors)
        avgAvgPrecisionRanking = average(avgPrecisionRankings)
        avgRankingLoss = average(rankingLosses)
        datasetEvaluation = {'coverageError': avgCoverageError, 'labelRankingAvgPrecision': avgAvgPrecisionRanking, 'labelRankingLoss': avgRankingLoss}
        overallEvaluation[dataset.name] = datasetEvaluation
    return overallEvaluation
```

```python
# OneVsRest classifiers

modelLogistic = OneVsRestClassifier(LogisticRegression(solver='sag'), n_jobs=-1)
modelKNN = OneVsRestClassifier(KNeighborsClassifier(n_neighbors=10, weights='distance'), n_jobs=-1)
modelSVC = OneVsRestClassifier(SVC(kernel='rbf', C=2), n_jobs=-1)
```

```python
evaluationLogistic = evaluationWithOtherModel([embeddingMIMIC_500, embeddingMIMIC_1000, bagOfWordsHDP, tfIdfHDP, bagOfWordsLDAasLSI], modelLogistic, 5)
pp.pprint(evaluationLogistic)
```

```python
evaluationKNN = evaluationWithOtherModel([embeddingMIMIC_500, embeddingMIMIC_1000, bagOfWordsHDP, tfIdfHDP, bagOfWordsLDAasLSI], modelKNN, 5)
pp.pprint(evaluationKNN)
```

```python
evaluationSVC = evaluationWithOtherModel([embeddingMIMIC_500, embeddingMIMIC_1000, bagOfWordsHDP, tfIdfHDP, bagOfWordsLDAasLSI], modelSVC, 5)
pp.pprint(evaluationSVC)
```
